# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.seasonal import STL
```

# Reading

```python
path = '../data/GOOGL_2006-01-01_to_2018-01-01.csv'
df = pd.read_csv(path)
df.head()
```

# Preprocessing

```python
df['Date'].iloc[0]
# Date is in string format
```

```python
# Converting Date column to datetime
date_format = '%Y-%m-%d'
df['Date'] = [ datetime.strptime(x, date_format) for x in df['Date']]
df.head()
```

```python
df['Name'].unique()
```

```python
df.drop('Name', axis=1, inplace=True)
```

```python
# Checking for null values
df.isna().sum()
```

# Visualization

```python
# Checking if data is equally spaced
difference = df['Date'].iloc[1:].reset_index(drop=True) - df['Date'].iloc[:-1].reset_index(drop=True)

plt.subplots(figsize=(10,3))
plt.scatter(df['Date'].iloc[1:], pd.DataFrame(difference)['Date'].dt.days, s=2)
plt.xlabel("Sample's date")
plt.ylabel('Days from last sample')
plt.title("Checking if data is equally spaced")
plt.show()
```

```python
# Distributions for different variables
for column in df.columns[1:-1]:
    aux_df = df[['Date', column]]
    plt.plot(aux_df['Date'], aux_df[column], label=str(column))

plt.ylabel("Currency")
plt.xlabel("Date")
plt.title("Distribution for Open, High, Low and Close")
plt.legend()
plt.show()
```

```python
plt.subplots(figsize=(10, 3))
plt.plot(df['Date'], df['Volume'])
plt.ylabel("Volume")
plt.xlabel("Date")
plt.title("Volume Distribution across the years")
plt.show()
```

```python
plt.subplots(figsize=(10, 3))

filter_1 = df[df['Date'].dt.year == 2010]
filter_2 = filter_1[filter_1['Date'].dt.month == 6]

for column in df.columns[1:-1]:
    aux_df = filter_2[['Date', column]]
    plt.plot(aux_df['Date'], aux_df[column], 'o-', label=str(column))

plt.ylabel("Currency")
plt.xlabel("Date")
plt.title("Distribution for Open, High, Low and Close for May 2010")
plt.legend()
plt.show()
```

# Extracting Trends

```python
for column in df.columns[1:]:
    aux_series = pd.Series(df[column])
    aux_series.index = df['Date']

    stl = STL(aux_series, period= 365)# 30-2*4) # 2 days * 4 weeks on each month
    res = stl.fit()
    fig = res.plot()
    plt.show()
```

```python
for column in df.columns[1:]:
    aux_series = pd.Series(df[column])
    aux_series.index = df['Date']

    stl = STL(aux_series, period= 365)# 30-2*4) # 2 days * 4 weeks on each month
    res = stl.fit()

    # Removing trends
    diff = pd.DataFrame(aux_series - res.trend, columns=['Diff']).reset_index()
    # Aggregating by month and days
    diff['Month'] = diff['Date'].dt.month
    diff['Day'] = diff['Date'].dt.day
    
    fig, ax = plt.subplots(1, 2, figsize = (10, 6))

    sns.boxplot(x='Diff', y='Month', data=diff, ax = ax[1], orient='h')
    ax[1].set_title('Boxplots Grouped by Month')

    sns.boxplot(x='Diff', y='Day', data=diff, ax = ax[0], orient='h')
    ax[0].set_title('Boxplots Grouped by Day')

    plt.suptitle("Checking seasonality for:"+column)
    plt.show()
```
